```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('company_sales_data.csv')
data
```

```python
month = data['month_number'].to_list()
sales = data['total_profit'].to_list()
```

```python
month
```

```python
graf = plt.plot(month, sales, linewidth = 3)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(month)
plt.yticks([x for x in range(min(sales), max(sales), 20000)])
plt.title('Sales last year')
plt.show()
```
